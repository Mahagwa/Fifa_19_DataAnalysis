## Countries with the most players
From the plot below, we see that the highest number of players come from England. In the top ten countries with the highest number of players, we see that six are from Europe with the top three being England, Germany and Spain. Argentina comes in fourth, Brazil sixth and Colombia eigth as the countries from South America. Then we have Japan in ninth as the only Asian country in the top ten. We see France, Italy and Netherlands as the other European countries in the list.


```python
#plotting top five countries with the highest number of players
df.Nationality.value_counts().head(10).plot(kind='bar',figsize=(10,8), xlabel='Country', ylabel='Count', title= 'Nationalities with Highest number of  players')
```

## Player Age Distribution
From the plot below derived from the data, we see a high number of players' ages are between the range of 19 and 28 years. The highest number of players being of age 21. We also see the youngest player being of age 16 and the oldest being above 41. Players aged 26 and 24 come at second and third highest respectively.

```python
#plotting the age distribution of players
df.Age.value_counts().sort_index(ascending=True).plot(kind='bar',figsize=(10,8), xlabel= 'Age', ylabel= 'Count', title= 'Player Age Count')
```

## Relationship between Age, Agility and Stamina
In the plot, we see a correlation between Stamina and Agility where players with lower agility also have lower stamina. We also see players above 35 generally having lower stamina. Players with higher Agility generally have higher Stamina too showing that the two are partially co-dependent variables.


```python
# plotting relationship between Age, Agilty and stamina
plt.figure(figsize = [11.69,9.27])
plt.scatter(data = df, x = 'Age', y = 'Agility',
                c = 'Stamina')
plt.colorbar(label = 'Stamina')
plt.xlim(10,48)
plt.ylim(5,100)
plt.xlabel('Age')
plt.ylabel('Agility')
plt.title('Relationship between Age, Agilty and stamina')
```
